## Write a Pandas program to display the first and last name and date of joining of the employees who is either Sales Representative or Sales Man.

```python
import pandas as pd
```

```python
def display_employees():
    df = pd.read_csv('Employees.csv')
    
    result = df[df['job_id'].isin(['SA_REP','SA_MAN'])]
    
    for (idx, row) in result.iterrows():
        print(row['first_name'], row['last_name'], row['job_id'], row['hire_date'])
```

```python
display_employees()
```
