## Write a Pandas program to display the movies (title, number of votes) that received specified number of votes.

```python
import pandas as pd
```

```python
def display_movies(n: int):
    df = pd.read_csv('movies_metadata.csv', usecols=['title', 'vote_count'])
    
    return df[df['vote_count'] >= n]
```

```python
display_movies(500)
```
