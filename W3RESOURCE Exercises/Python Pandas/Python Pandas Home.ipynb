# Pandas Data Structure

## Key and Imports

```python
import pandas as pd
```

```python
def create_dataframe(d: dict):
    df = pd.DataFrame(d)
    return df
```

```python
create_dataframe({'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]})
```

```python
def create_series(l: list):
    s = pd.Series(l)
    return s
```

```python
create_series([2, 4, 6, 8, 10])
```
